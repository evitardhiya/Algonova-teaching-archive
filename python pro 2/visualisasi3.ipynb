```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('menu.csv')
df.head(5)
```

```python
df.info()
```

```python
#kategoru mana yang memiliki rata-rata gula yang paling tinggi?
df.groupby('Category')['Sugars'].mean().sort_values().plot(kind='barh')
```

```python
#manakah menu yang paling banyak diminati?
df['Category'].value_counts().plot(kind='pie')
```

```python
#item manakah yang memiliki protein paling tinggi?
df.groupby(['Category','Item'])['Protein'].max().sort_values(ascending=False).head(5)
```

```python
#item manakah yang memiliki sugars paling rendah?
df.groupby(['Category','Item'])['Sugars'].min().sort_values(ascending=True).head(5)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Total Fat', y='Cholesterol')
plt.title('Relasi Total Fat vs Cholesterol')
plt.xlabel('Total Fat (g)')
plt.ylabel('Cholesterol (mg)')
plt.grid(True)
plt.show()

correlation = df['Total Fat'].corr(df['Cholesterol'])
print(f'Korelasi antara Total Fat dan Cholesterol: {correlation:.2f}')
```
