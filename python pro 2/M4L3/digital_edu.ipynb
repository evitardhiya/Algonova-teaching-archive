- langs — daftar bahasa yang digunakan pengguna

- people_main -karakteristik yang utama bagi orang. (1 - kecerdasan dan kreativitas; 2 - kebaikan dan kejujuran; 3 - kecantikan dan kesehatan; 4 - kekuasaan dan kekayaan; 5 - keberanian dan ketekunan; 6 - humor dan cinta hidup).
- career_start — tahun mulai bekerja.
- career_end — tahun berakhirnya pekerjaan.
- result — apakah pengguna membeli kursus pemrograman (0 —tidak, 1 — ya).


```python
df_test = pd.read_csv('test.csv')
df_train = pd.read_csv('train.csv')
```

```python
df_train.drop(['bdate', 'has_photo', 'id', 'followers_count', 'langs', 'last_seen', 'city', 'career_start', 'career_end', 'occupation_name'], axis=1, inplace=True)
df_test.drop(['bdate', 'has_photo', 'followers_count', 'langs', 'last_seen', 'city', 'career_start', 'career_end', 'occupation_name'], axis=1, inplace=True)
```

```python
df_train.info()
```

```python
df_train['education_form'].unique()
```

```python
def fill_education_form(education):
    if education == 'Full-time':
        return 1
    elif education == 'Distance Learning':
        return 2
    elif education == 'Part-time':
        return 3
    else:
        return 0
 
df_train['education_form'] = df_train['education_form'].apply(fill_education_form)
df_test['education_form'] = df_test['education_form'].apply(fill_education_form)
```

```python
df_train['education_status'].unique()
```

```python
def fill_education_status(status):
    if status == 'Student (Specialist)':
        return 1
    elif status == 'Alumnus (Specialist)':
        return 2
    elif status == 'PhD':
        return 3
    elif status == "Student (Bachelor's)":
        return 4
    elif status == "Alumnus (Master's)":
        return 5
    elif status == "Alumnus (Bachelor's)":
        return 6
    elif status == "Student (Master's)":
        return 7
    elif status == 'Undergraduate applicant':
        return 8
    elif status == 'Candidate of Sciences':
        return 9
    else:
        return 0

df_train['education_status'] = df_train['education_status'].apply(fill_education_status)
df_test['education_status'] = df_test['education_status'].apply(fill_education_status)
```

```python
df_train['occupation_type'].unique()
```

```python
def fill_occupation_type(x):
    if x == 'university':
        return 1
    elif x == 'work':
        return 2
    else:
        return 0
 
df_train['occupation_type'] = df_train['occupation_type'].apply(fill_occupation_type)
df_test['occupation_type'] = df_test['occupation_type'].apply(fill_occupation_type)
```

```python
def fix_grad(data):
    nilai_tengah = data[(data > 1900) & (data < 3000)].median()
    return data.apply(lambda x: nilai_tengah if x <= 1900 or x >= 3000 else x)
df_train['graduation'] = fix_grad(df_train['graduation'])
df_test['graduation'] = fix_grad(df_test['graduation'])
```

```python
convert_dtype = ['has_mobile', 'graduation', 'relation']
df_train[convert_dtype] = df_train[convert_dtype].astype(int)
df_test[convert_dtype] = df_test[convert_dtype].astype(int)
```

```python
def clean_main_value(val):
    if val == False or val == 'False':  
        return 0
    try:
        return int(val)
    except:
        return 0  

df_train['life_main'] = df_train['life_main'].apply(clean_main_value)
df_test['life_main'] = df_test['life_main'].apply(clean_main_value)
df_train['people_main'] = df_train['people_main'].apply(clean_main_value)
df_test['people_main'] = df_test['people_main'].apply(clean_main_value)
```

-----------Model machine learning---------------

```python
#1. import modul sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
#2. bagi data menjadi Data variabel(X) dan target(Y)
X_train = df_train.drop('result', axis = 1)
y_train = df_train['result']
X_test = df_test.drop('id', axis = 1)
```

```python
#3. membuat data menjadi Nilai standar
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#4. Membuat variabel model KNN dengan k=5
classifier1 = KNeighborsClassifier(n_neighbors = 5)
#4. Membuat variabel model KNN dengan k=3
classifier2 = KNeighborsClassifier(n_neighbors = 3)

#5. melakukan training model
classifier1.fit(X_train, y_train)
classifier2.fit(X_train, y_train)
```

```python
#6. Melakukan prediksi pada data Test
y_pred1 = classifier1.predict(X_test)
y_pred2 = classifier2.predict(X_test)
```

```python
result = pd.DataFrame({'id' : df_test['id'], 'result': y_pred1})
result.to_csv('result_digitalEdu.csv', index = False)
```
